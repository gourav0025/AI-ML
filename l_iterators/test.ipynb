```python

class Seq_iterator:

    def __init__(self, seq):
        self._seq = seq
        self._index = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if(self._index < len(self._seq)):
            item = self._seq[self._index]
            self._index += 1
            return item
        else:
            raise StopIteration
        
lst = [1,2,3,4,5]

custom_iter = Seq_iterator(lst)

for i in custom_iter:
    print(i)

```
