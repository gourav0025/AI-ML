```python

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Vector(self.x+other.x, self.y+other.y)
    
    def __str__(self):
        return f"x = {self.x}, y = {self.y}"
```

```python
v1 = Vector(2,3)
v2 = Vector(4,2)

resV = v1+v2
print(resV)
```
