```python
## Data serialization
import json
data = {
    'name' : 'Krish', 'age': 25
}

json_str = json.dumps(data)   # converted to json string
print(json_str, type(json_str))

json_parsed = json.loads(json_str) # converting string back into dictionary
print(json_parsed, type(json_parsed))
```

```python
# csv

import csv

with open('example.csv', mode='w+') as file:
    writer = csv.writer(file)
    writer.writerow( ['name', 'age'])
    writer.writerow( ['Gourav', 21])

with open('example.csv', mode='r') as file:
    reader=csv.reader(file)

    for row in reader:
        print(row)
```
