# Numpy array basics

## Creating array from List

```python
arr_1d = np.array([1,2,3,4,5])
print("1d array:", arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]])
print(arr_2d)
```

## List Vs Numpy array

#### Scaler Multiplication difference

```python
py_list = [1,2,3,4]
print(py_list*2)

np_array = np.array([1,2,3,4])
print(np_array*2)
```

#### Time consumption difference

```python
import time
start = time.time()
py_list = [i*2 for i in range(10000000)]
print("List time taken:", time.time()-start)

start = time.time()
np_list = np.arange(10000000)*2
print("Array time taken:", time.time()-start)
```

## Creating different arrays from scratch

#### Creates n-d array of all elements zero

```python
## 1-d array with 'n' elements
zero_arr = np.zeros((3))  
print(zero_arr)

## 2-d array with 'n*m' elements
zero_arr = np.zeros((3,4))  
print(zero_arr)

## 3-d array with 'n*m*k' elements
zero_arr = np.zeros((3,4,2))  
print(zero_arr)
```

####  Create n-d array of all elements with value 1

```python
ones_arr = np.ones((2,3))  
print(ones_arr)
```

#### Creating n-d array of a constant value

```python
k_arr = np.full((2,3), 7) # ((dimensions tuple), constant)
print(k_arr)
```

#### Creating random valued arrays

```python
# random values in the array are between 0 to
rand = np.random.random((2,3))
print(rand)

# Getting single digit random values
rand =(np.random.random((2,3))*10)//1
print(rand)

# Getting double digit random values
rand =(np.random.random((2,3))*100)//1
print(rand)
```

# Properties of an Array

```python
arr = np.array([[1,2,3],
                [4,5,6]
                ])

print("Shape", type(arr.shape), arr.shape)
print("Dimension", type(arr.ndim), arr.ndim)
print("Size", type(arr.size), arr.size)   # total number of elements
print("Dtype", type(arr.dtype),arr.dtype)

```

# Array Reshaping 

#### reshape(), flatten(), ravel(), Transpose

```python
arr = np.arange(12)
print("Original array", arr , "\n", "-"*110)

# New shape must have same total number of elements as original element

a = arr.reshape((3,4))
b = arr.reshape((2,2,3))
print(a)
print(b)
print("-"*110)

## Using -1 instead of row or col, making numpy automatically calculate for it

c = arr.reshape(6,-1)      # cols computed = 2
d = arr.reshape(-1, 4)     # rows computed = 3
print(c)
print(d)
print("-"*110)

# Flattening, convertes n-d array to 1-d (returns copy)
flat_1 = a.flatten()
flat_2 = b.flatten()
print(flat_1)
print(flat_2)
print("-"*110)


# Ravel, converts n-d array to 1-d but returns view of actual array
## thus modifying returned array modifies the actual array

rav_1 = a.ravel()
rav_1[0] = 2
print(rav_1, a)
print("-"*110)

# Transpose
a = np.array([
    [1,2,3],
    [4,5,6]
])

print(a.T)
```
