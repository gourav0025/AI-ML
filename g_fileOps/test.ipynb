```python
## Read the whole file

with open("example.txt", "r") as file:  
    print(file, type(file))
    content = file.read()

    print(content)
```

```python
# Iterating over file object

with open("example.txt", "r") as f:
    for line in f:
        print(line.strip())

```

```python
# Writing to the file
with open("example.txt", 'w') as f:
    f.write("overwritten")
```

```python
# Appending in the file
with open("example.txt", 'a') as f:
    f.write("\n")
    f.write("New content added")
```

```python
# Write/Append a List of lines to a file
lines = ['\n', 'hey there\n', 'I am using whatsapp']

with open("example.txt", "a") as f:
    f.writelines(lines)
```

```python
# Binary File

## Writing to a binary file

data = b'\x00\x01\x02\x03\x04' # written in hexadecimal  # Equivalent to 0, 1, 2, 3, 4 in decimal and 00000000, 00000001, 00000010, 00000011, 00000100 in bytes.
with open('exp.bin', 'wb') as file:
    file.write(data)
```

```python
## Reading from a binary file
with open('exp.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
# Reading from and writing to dest file
with open("example.txt", "r") as f:
    content = f.read()

with open("dest.txt", "w") as f:
    f.write(content)
```

```python
# Count number of lines, words and Characters in a file

with open("example.txt", "r") as f:
    lines = f.readlines()
    line_count = len(lines)
    word_count = sum(len(line.split()) for line in lines) 
    char_count = sum(len(line) for line in lines)

print(line_count, word_count ,char_count )
    

    
```
