```python
class Car:
    wheels = 4
    def greet(self):
        print("Hi")

    def __init__(self, name):
        self.__name = name

    @staticmethod
    def isRunning():
        print("whoom whoom")


c1 = Car()
c1.name = "audi"
c1.greet()
c1.isRunning()
Car.greet()
```
